```python
import time

# Define the time interval in seconds (3600 seconds = 1 hour)
interval = 800

# Number of times to run the code
num_executions = 100

# Counter to keep track of executions
execution_count = 0

while execution_count < num_executions:
    # Your code to be executed every hour goes here
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find_all('table')[12]
    table2 =table.find_all('table')[0]
    table3 = table2.find_all('table')[0]
    table4 = table3.find_all('table')[2]
    df = pd.DataFrame(columns = titles)
    column_data = table4.find_all('tr')
    for row in column_data[1:]:
      row_data = row.find_all('td')
      individual_row = [data.text.strip() for data in row_data]
      #print(individual_row[1:])
      df.loc[len(df)] = individual_row[1:]
    df = df[df['Headlines'] !='Loading…']
    df.loc[df['Time'].str.contains(r'^[A-Za-z]', regex=True), 'Time'] = datetime.now() - timedelta(days=1)
    df['Time']=pd.to_datetime(df['Time'])
    complete = pd.read_csv(os.path.join(DIR_DATA, 'headlines.csv'))[['Time','Headlines']]
    merged_df = pd.concat([complete, df]).drop_duplicates(subset=['Headlines'])
    merged_df['Time']=pd.to_datetime(merged_df['Time'])
    merged_df = merged_df.sort_values(by = 'Time').reset_index(drop=True)
    merged_df.to_csv(os.path.join(DIR_DATA, 'headlines.csv'))

    print("This code runs every hour. Execution:", execution_count + 1)
    print(merged_df)

    # Increment the execution count
    execution_count += 1

    # Sleep for the specified interval
    time.sleep(interval)
```
